```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(t):
    return (1-t) / t
```

```python
def bound(N, m, alpha, g, kappa, t):
    A_1 = np.log(1 + g / np.sqrt(2 * F(t) * m ** (1 - 2 * alpha)))
    A_2 = 0.5 * np.log(1 / F(t))
    A_3 = 0.5 * np.log(m ** (1 - 2*alpha) / kappa **2) - 0.5
    A = A_1 + A_2 + A_3
    
    B_1 = kappa**2 / t * (1 + g**2 / N)
    B_2 = g**2 / N * (1+ 1 / F(t))
    B = B_1 + B_2
    
    return N / m**alpha * A + N / m ** (1-alpha) * B
```

```python
t = np.logspace(-3, -0.01, 100)
```

```python
N = 100
m = 10000
alpha = 0.3
g = 10
kappa = 0.5
```

```python
plt.loglog(t, bound(N, m, alpha, g, 0.5, t), label=fr"$\kappa=${0.5:.1f}")
plt.loglog(t, bound(N, m, alpha, g, 1.0, t),label=fr"$\kappa=${1.0:.1f}")
plt.loglog(t, bound(N, m, alpha, g, 2.0, t),label=fr"$\kappa=${2.0:.1f}")

plt.xlabel("t")
plt.legend()
plt.savefig("bound-t.png", dpi=400)
```
