```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle

from matplotlib.colors import LogNorm
import seaborn as sns
```

```python
path = "../classification-bounds/emotions-heatmap-relaxed.pkl"
with open(path, "rb") as file:
    results = pickle.load(file)
```

```python
df = pd.DataFrame(results)
df = df[(df.alpha <= 0.5) & (df.t >= 0.5)]
dd = df.groupby(["alpha", "t"]).min().reset_index()
piv = dd.pivot("alpha", "t")
```

```python
fig = plt.figure(figsize=(12, 10))
ax = plt.gca()
sns.heatmap(
    piv["loss_Q"],
    annot=True,
    fmt="1.1e",
    cmap="Blues",
    norm=LogNorm(vmin=piv["loss_Q"].values.min(), vmax=piv["loss_Q"].values.max()),
    ax=ax,
)
plt.xlabel(r"$t$")
plt.ylabel(r"$\alpha$")
plt.show()
fig.savefig("Relaxed-loss-Q.png", dpi=400, bbox_inches='tight')
```
