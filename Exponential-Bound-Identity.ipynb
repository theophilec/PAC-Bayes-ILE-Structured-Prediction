```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bound(x, a=1):
    e = np.exp(1)
    return a * e / (e - 1) * (1 - np.exp(- x / a))
```

```python
x = np.linspace(0, 1, 100)
```

```python
a_t = np.logspace(0, 2, 5)
```

```python
for a_ in a_t:
    plt.plot(x,bound(x, a=a_), label=f"Bound for a= {a_:.0f}")
plt.plot(x,x, label="Identity")
plt.xlabel("x")
plt.legend()
plt.savefig('exponential-bound-identity.png', dpi=400)
```
